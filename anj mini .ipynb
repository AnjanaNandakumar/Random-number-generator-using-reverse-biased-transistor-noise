```python
from matplotlib import pyplot
import numpy as np
f = open(r"C:\Users\ANJANA NANDAKUMAR\Downloads\mini testuuuu.txt", "r")
k=[]
for line in f:
    k.append(line.strip())
k.remove('')

x = [k[i:i + 8] for i in range(0, len(k), 8)] 
k = list(map(int, k))
r=[]
for i in range(0,len(x)):
    r.append(str("".join(x[i])))
for i in range(0,len(x)):
    r[i]=int(r[i],2)
```

```python
x = np.array(r[:1758])
y = np.array(r[1758:])
print("Correlation of the Raw Data")
print(np.corrcoef(x,y))

pyplot.scatter(x, y)
pyplot.show()
```
